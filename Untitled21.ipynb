```python
import requests
import pandas as pd

# رابط API للإطلاقات
url = "https://api.spacexdata.com/v4/launches/past"

# جلب البيانات
response = requests.get(url)
data = response.json()

# تحويل البيانات إلى DataFrame
df = pd.json_normalize(data)

print("عدد الصفوف والأعمدة:", df.shape)
df.head(3)
```

```python
# اختيار أعمدة رئيسية (ممكن نضيف/نحذف لاحقًا حسب التحليل)
cols = [
    'name',               # اسم المهمة
    'date_utc',           # تاريخ الإطلاق
    'success',            # نجاح المهمة
    'rocket',             # معرف الصاروخ
    'launchpad',          # معرف منصة الإطلاق
    'payloads',           # قائمة المعززات/الحمولات
]

df = df[cols]
df.head()
```

```python
# تحويل تاريخ الإطلاق إلى نوع تاريخ
df['date_utc'] = pd.to_datetime(df['date_utc'])

# إعادة تسمية الأعمدة لتكون أوضح
df.rename(columns={
    'name': 'mission_name',
    'success': 'landing_success',
}, inplace=True)

# عرض أول صفوف
df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(font_scale=1.1)  # إعداد جمالي بسيط ومتوافق

# تأكيد الأعمدة الحالية
print(df.columns.tolist())
print(df.shape)
df.head(3)
```

```python
# استخراج أول قيمة landing_success من قائمة cores لكل إطلاق
import numpy as np

def first_landing_success(row):
    cores = row.get('cores', [])
    if isinstance(cores, list) and len(cores) > 0 and isinstance(cores[0], dict):
        return cores[0].get('landing_success')
    return np.nan

# لو df الأصلي ما فيه 'cores' لأننا قصصنا الأعمدة، نرجع نقرأ الإطلاقات مرة ثانية ونلصق العمود فقط
if 'cores' not in df.columns:
    import requests
    launches_full = pd.json_normalize(requests.get("https://api.spacexdata.com/v4/launches/past").json())
    df = df.merge(
        launches_full[['id','cores']],
        left_on='mission_name', right_on=launches_full['name'] if 'name' in launches_full.columns else 'id',
        how='left'
    )
    # قد تكون عملية الدمج حسب اسم المهمة غير مثالية؛ الأفضل نعيد التطبيع من المصدر كامل لو ما ظبطت

# الأسلم: نعيد جلب df من المصدر الكامل ثم نعيد اختيار ودمج الأعمدة (بديل موصى به)
import requests
data = requests.get("https://api.spacexdata.com/v4/launches/past").json()
full = pd.json_normalize(data)

# نبني df جديد متكامل
df = full[['name','date_utc','success','rocket','launchpad','payloads','cores']].copy()
df.rename(columns={'name':'mission_name'}, inplace=True)
df['date_utc'] = pd.to_datetime(df['date_utc'])
df['landing_success'] = df.apply(first_landing_success, axis=1)

# دمج أسماء الصواريخ ومنصات الإطلاق كما فعلنا سابقًا
rockets = pd.json_normalize(requests.get("https://api.spacexdata.com/v4/rockets").json())[['id','name','mass.kg']]
rockets.rename(columns={'id':'rocket', 'name':'rocket_name', 'mass.kg':'rocket_mass_kg'}, inplace=True)

launchpads = pd.json_normalize(requests.get("https://api.spacexdata.com/v4/launchpads").json())[['id','name','locality']]
launchpads.rename(columns={'id':'launchpad','name':'launchpad_name','locality':'launchpad_location'}, inplace=True)

df = df.merge(rockets, on='rocket', how='left')
df = df.merge(launchpads, on='launchpad', how='left')

df.head(3)
```

```python
import requests
payloads = pd.json_normalize(requests.get("https://api.spacexdata.com/v4/payloads").json())

# أعمدة مفيدة من الحمولات
payloads_small = payloads[['id','mass_kg','orbit']].copy()

# دالة تجمع كتلة الحمولة وتختار المدار الأكثر تكرارًا لكل إطلاق
def summarize_payloads(payload_ids):
    if not isinstance(payload_ids, list) or len(payload_ids) == 0:
        return pd.Series({'payload_mass_sum_kg': np.nan, 'dominant_orbit': np.nan})
    sub = payloads_small[payloads_small['id'].isin(payload_ids)]
    mass_sum = sub['mass_kg'].sum(min_count=1)  # NaN إذا كلها NaN
    dom_orbit = sub['orbit'].mode().iloc[0] if not sub['orbit'].dropna().empty else np.nan
    return pd.Series({'payload_mass_sum_kg': mass_sum, 'dominant_orbit': dom_orbit})

payload_summary = df['payloads'].apply(summarize_payloads)
df = pd.concat([df, payload_summary], axis=1)

# أعمدة نهائية نظيفة
eda_cols = [
    'mission_name','date_utc','landing_success','success',
    'rocket_name','rocket_mass_kg',
    'launchpad_name','launchpad_location',
    'payload_mass_sum_kg','dominant_orbit'
]
df_eda = df[eda_cols].copy()

df_eda.head(5)
```

```python
# سنة الإطلاق
df_eda['year'] = df_eda['date_utc'].dt.year

# تحويل landing_success إلى 0/1 (والتعامل مع NaN)
df_eda['landing_success_bin'] = df_eda['landing_success'].map({True:1, False:0})
# بإمكانك إسقاط الصفوف اللي بدون هبوط (NaN) إذا تبين تحللين الهبوط فقط:
df_ls = df_eda.dropna(subset=['landing_success_bin']).copy()

# لمحة سريعة
print("صفوف بعد إسقاط NaN في landing_success:", df_ls.shape)
df_ls[['landing_success_bin','success','dominant_orbit']].head()
```

```python
# معدل نجاح الهبوط الكلّي
overall_rate = df_ls['landing_success_bin'].mean()
print("معدل نجاح الهبوط الإجمالي:", round(overall_rate*100, 2), "%")

# أعلى 5 مواقع إطلاق من حيث معدل نجاح الهبوط
site_rates = df_ls.groupby('launchpad_name')['landing_success_bin'].mean().sort_values(ascending=False).head(5)
print("\nأعلى مواقع من حيث النجاح:\n", (site_rates*100).round(1).astype(str) + "%")

# متوسط الكتلة حسب النجاح
mass_success = df_ls.groupby('landing_success_bin')['payload_mass_sum_kg'].mean().round(2)
print("\nمتوسط إجمالي كتلة الحمولة حسب النجاح (0=فشل،1=نجاح):\n", mass_success)

# معدل النجاح حسب المدار
orbit_rates = df_ls.groupby('dominant_orbit')['landing_success_bin'].mean().sort_values(ascending=False)
print("\nمعدل النجاح حسب المدار:\n", (orbit_rates*100).round(1).astype(str) + "%")
```

```python
import sqlite3

conn = sqlite3.connect("spacex.db")
df_ls.to_sql("launches", conn, if_exists="replace", index=False)

# مثال: معدل النجاح حسب المنصة
q1 = """
SELECT launchpad_name, ROUND(AVG(landing_success_bin)*100,1) AS success_rate_pct, COUNT(*) AS n
FROM launches
GROUP BY launchpad_name
ORDER BY success_rate_pct DESC
LIMIT 10;
"""
sql_site = pd.read_sql(q1, conn)
sql_site

# مثال: معدل النجاح حسب المدار
q2 = """
SELECT dominant_orbit, ROUND(AVG(landing_success_bin)*100,1) AS success_rate_pct, COUNT(*) AS n
FROM launches
GROUP BY dominant_orbit
ORDER BY success_rate_pct DESC;
"""
sql_orbit = pd.read_sql(q2, conn)
sql_orbit
```

```python
from pandas.plotting import table

def save_table_as_image(df_table, path, max_rows=12):
    import matplotlib.pyplot as plt
    fig, ax = plt.subplots(figsize=(10, 0.5 + 0.3*min(len(df_table), max_rows)))
    ax.axis('off')
    table(ax, df_table.head(max_rows), loc='center')
    plt.tight_layout()
    plt.savefig(path, dpi=200, bbox_inches='tight')
    plt.close(fig)

save_table_as_image(sql_site, "reports/tables/sql_site.png")
save_table_as_image(sql_orbit, "reports/tables/sql_orbit.png")
```

```python
# === Folium Map ===
import pandas as pd, numpy as np, requests, folium, os

# تأكد من وجود df_ls (نتائج landing_success_bin)
assert 'landing_success_bin' in df_ls.columns and 'launchpad' in df_ls.columns

# نجلب معلومات إحداثيات منصات الإطلاق (lat/lon)
launchpads_full = pd.json_normalize(requests.get("https://api.spacexdata.com/v4/launchpads").json())
lp = launchpads_full[['id','name','locality','latitude','longitude']].copy()
lp.rename(columns={'id':'launchpad','name':'launchpad_name','locality':'launchpad_location'}, inplace=True)

# معدل النجاح لكل منصة + عدد الإطلاقات
site_stats = (df_ls
              .groupby('launchpad')
              .agg(success_rate=('landing_success_bin','mean'),
                   n=('landing_success_bin','size'))
              .reset_index())
site_stats = site_stats.merge(lp, on='launchpad', how='left')

# مركز الخريطة = متوسط الإحداثيات المتاحة
center_lat = site_stats['latitude'].mean()
center_lon = site_stats['longitude'].mean()

m = folium.Map(location=[center_lat, center_lon], zoom_start=2, tiles="OpenStreetMap")

# مقياس اللون حسب النجاح
def colorize(rate):
    if pd.isna(rate): return "gray"
    if rate >= 0.8: return "green"
    if rate >= 0.6: return "orange"
    return "red"

for _, r in site_stats.dropna(subset=['latitude','longitude']).iterrows():
    rate_pct = round(r['success_rate']*100,1) if pd.notna(r['success_rate']) else None
    popup = folium.Popup(html=f"""
    <b>{r['launchpad_name']}</b><br>
    الموقع: {r.get('launchpad_location','')}<br>
    معدل النجاح: {rate_pct}%<br>
    عدد الإطلاقات: {int(r['n'])}
    """, max_width=250)
    folium.CircleMarker(
        location=[r['latitude'], r['longitude']],
        radius=8 + (r['n']**0.5),  # تكبير بسيط حسب العدد
        color=colorize(r['success_rate']),
        fill=True, fill_opacity=0.7,
        popup=popup,
        tooltip=f"{r['launchpad_name']} — {rate_pct}%"
    ).add_to(m)

os.makedirs("reports", exist_ok=True)
m.save("reports/map.html")
print("تم حفظ الخريطة التفاعلية: reports/map.html — خذي لقطة شاشة لها للعرض.")
```

```python
# === التصنيف (Predictive Classification) ===
import pandas as pd, numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, roc_auc_score, ConfusionMatrixDisplay, RocCurveDisplay)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt, os

# نحضّر الداتا: نسقط الصفوف اللي ما فيها الهدف أو الميزات
clf_df = df_ls[['landing_success_bin','year','payload_mass_sum_kg','launchpad_name','rocket_name','dominant_orbit']].copy()
clf_df = clf_df.dropna(subset=['landing_success_bin'])

# نعالج NaN في الميزات الرقمية بتبديلها بمتوسط (بسيط)
for c in ['payload_mass_sum_kg','year']:
    clf_df[c] = clf_df[c].astype(float)
    if clf_df[c].isna().any():
        clf_df[c].fillna(clf_df[c].median(), inplace=True)

X = clf_df[['year','payload_mass_sum_kg','launchpad_name','rocket_name','dominant_orbit']]
y = clf_df['landing_success_bin'].astype(int)

# أعمدة
num_cols = ['year','payload_mass_sum_kg']
cat_cols = ['launchpad_name','rocket_name','dominant_orbit']

preprocess = ColumnTransformer(
    transformers=[
        ('num','passthrough', num_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ])

# نموذجين للتجربة
models = {
    "Logistic": LogisticRegression(max_iter=200, class_weight='balanced'),
    "RandomForest": RandomForestClassifier(n_estimators=300, random_state=42, class_weight='balanced')
}

# تقسيم
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

results = {}
os.makedirs("reports/figures", exist_ok=True)

for name, clf in models.items():
    pipe = Pipeline(steps=[('prep', preprocess),
                          ('clf', clf)])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    y_proba = pipe.predict_proba(X_test)[:,1] if hasattr(pipe, "predict_proba") else None

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    auc = roc_auc_score(y_test, y_proba) if y_proba is not None else np.nan

    results[name] = dict(Accuracy=acc, Precision=prec, Recall=rec, F1=f1, AUC=auc)

    # مصفوفة الالتباس
    disp = ConfusionMatrixDisplay.from_estimator(pipe, X_test, y_test)
    plt.title(f"Confusion Matrix — {name}")
    plt.tight_layout()
    plt.savefig(f"reports/figures/confusion_matrix_{name}.png", dpi=200)
    plt.show()

    # ROC
    if y_proba is not None:
        RocCurveDisplay.from_predictions(y_test, y_proba)
        plt.title(f"ROC Curve — {name} (AUC={auc:.3f})")
        plt.tight_layout()
        plt.savefig(f"reports/figures/roc_{name}.png", dpi=200)
        plt.show()

print("المقاييس:")
for k,v in results.items():
    print(k, {m: round(val,3) if pd.notna(val) else None for m,val in v.items()})
```
